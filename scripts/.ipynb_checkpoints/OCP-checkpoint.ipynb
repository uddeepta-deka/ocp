```python
import numpy as np
import numpy.random as npr
import random as rd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
alpha = 1
beta = 1
a = 1.0 #[-a,a] denotes the interval
mu = -200.
N = 100 #number of particles/eigenvalues
nr = 100000 #number of realizations
```

```python
def en(x):
    'Calculating the energy for a given set x'
    
    count = 0 #it counts the number of eigenvalues within the interval
    Ex = 0 
    for i in range(N):
        Ex = Ex - 2.*alpha*(2*i-N-1)*x[i] + N * 0.5 * (x[i]**2)
        if -a <= x[i] <= a:
            count+=1.0
    Ex = Ex + (mu * count)
    return Ex
```

```python
def x_calc():
    xi = npr.normal(0.,1.,N)  #initial configuration
    eni = en(xi)
    st =[]
    x = []
    e = []
    AccRejRatio = []
    accept = 0
    'Monte-Carlo step: find the config after nr number of realizations'    
    for step in range(nr):
        xn = xi.copy()
        loc = npr.randint(0,N) #generates location of the eigenvalue to be changed in next iteration
        xn[loc] = xn[loc] + npr.uniform(-1/np.sqrt(N),1/np.sqrt(N)) #new configuration
        enn = en(xn)
        dE = enn-eni
        
        #Metropolis-Hastings step
        r = rd.random()
        if np.log(r)<min(0,-beta*dE):
            xi = xn
            eni=en(xi)
            accept+= 1
        
        
        if (step+1) % (2*N)==0:
            st.append(step+1)
            x.append(xi)
            e.append(eni)
            AccRejRatio.append(accept/(step+1))
            
    return st,x,e,AccRejRatio
```

```python
step, config, energy, Acc_Rej = x_calc()
```

```python
plt.subplot(121)
plt.plot(step,energy)
plt.subplot(122)
plt.plot(step,Acc_Rej)
plt.xlabel('step')
```

```python
#considering the burn-in time
del(config[0:300])
print(len(config))
```

```python
x_flat = [item for sublist in config for item in sublist]
```

```python
sns.set(color_codes=True)
sns.set(rc={'figure.figsize':(8.,5.)})

ax = sns.distplot(x_flat, kde=True)
ax.set(xlabel = 'x',ylabel = r'$\rho(x)$')
fig = ax.get_figure()
fig.suptitle(r'$\mu = 0, N = 100, a = 1, \alpha = 1, \beta = 1$', fontsize=12)
plt.axvline(1.0, color = 'r')
plt.axvline(-1.0, color = 'r')
```
